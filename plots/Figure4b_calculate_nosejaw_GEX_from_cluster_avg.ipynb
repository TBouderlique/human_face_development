```r
hvg <- read.csv("highly_variable_v1.csv")
spatial <- read.csv("master_spatial_n2.csv")
```

```r
snpscore <- read.csv("genes_with_SNP_scores_full.csv")
```

```r
jaw_ex_snp <- jaw_ex[rownames(jaw_ex) %in% snpscore$gene, ]
nose_ex_snp <- nose_ex[rownames(nose_ex) %in% snpscore$gene, ]
```

```r
genes_in_spatial <- unique(unlist(spatial))
genes_in_hvg <- hvg[[1]]
```

```r
jaw_ex_hvg <- jaw_ex[rownames(jaw_ex) %in% genes_in_hvg, ]
jaw_ex_spatial <- jaw_ex[rownames(jaw_ex) %in% genes_in_spatial, ]

nose_ex_hvg <- nose_ex[rownames(nose_ex) %in% genes_in_hvg, ]
nose_ex_spatial <- nose_ex[rownames(nose_ex) %in% genes_in_spatial, ]
```

```r
write.csv(jaw_ex_hvg, 'jaw_ex_hvg.csv')
write.csv(nose_ex_hvg, 'nose_ex_hvg.csv')
write.csv(jaw_ex_spatial, 'jaw_ex_spatial.csv')
write.csv(nose_ex_spatial, 'nose_ex_spatial.csv')
```

```r
write.csv(jaw_ex_snp, 'jaw_ex_snp.csv')
write.csv(nose_ex_snp, 'nose_ex_snp.csv')
```

```r
nose_mat <- as.matrix(sapply(nose_ex_snp, as.numeric))
jaw_mat  <- as.matrix(sapply(jaw_ex_snp, as.numeric))
nose_jaw_snp_diff <- nose_mat - jaw_mat
rownames(nose_jaw_snp_diff)  <- rownames(jaw_ex_snp)
avg_score_all <- rowMeans(nose_jaw_snp_diff, na.rm = TRUE)

nose_jaw_snp_diff_df <- as.data.frame(nose_jaw_snp_diff)
nose_jaw_snp_diff_df$avg_score_all <- avg_score_all
write.csv(nose_jaw_snp_diff_df, 'nose_jaw_full_diff_df.csv')
```

```r
nose_mat <- as.matrix(sapply(nose_ex_spatial, as.numeric))
jaw_mat  <- as.matrix(sapply(jaw_ex_spatial, as.numeric))
```

```r
nose_jaw_spatial_diff <- nose_mat - jaw_mat
```

```r
rownames(nose_mat) <- rownames(nose_ex_spatial)
rownames(jaw_mat)  <- rownames(jaw_ex_spatial)
```

```r
rownames(nose_jaw_spatial_diff)  <- rownames(jaw_ex_spatial)
```

```r
nose_mat <- as.matrix(sapply(nose_ex_hvg, as.numeric))
jaw_mat  <- as.matrix(sapply(jaw_ex_hvg, as.numeric))
```

```r
nose_jaw_hvg_diff <- nose_mat - jaw_mat
```

```r
rownames(nose_mat) <- rownames(nose_ex_hvg)
rownames(jaw_mat)  <- rownames(jaw_ex_hvg)
rownames(nose_jaw_hvg_diff)  <- rownames(jaw_ex_hvg)
```

```r
# Calculate row means (average across cell types)
avg_score_hvg <- rowMeans(nose_jaw_hvg_diff, na.rm = TRUE)
avg_score_spatial <- rowMeans(nose_jaw_spatial_diff, na.rm = TRUE)
```

```r
nose_jaw_hvg_diff_df <- as.data.frame(nose_jaw_hvg_diff)
nose_jaw_hvg_diff_df$avg_score <- avg_score_hvg

nose_jaw_spatial_diff_df <- as.data.frame(nose_jaw_spatial_diff)
nose_jaw_spatial_diff_df$avg_score <- avg_score_spatial
```
