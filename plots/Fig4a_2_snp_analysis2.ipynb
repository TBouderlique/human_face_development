```r
segment_genes <- read.csv('segment_genes_500k.csv')
```

```r
cluster_expr <- read.csv("white_gene_cluster55_averages_july2025.csv", row.names = 1)
```

```r


# For each segment, calculate average expression per cluster of mapped genes
# (Assuming segment_genes has columns 'Segment' and 'genes', and genes is a comma-separated string)

segment_cluster_averages <- segment_genes %>%
  rowwise() %>%
  mutate(
    avg_expr = list({
      gene_list <- unlist(strsplit(genes, ","))
      genes_in_expr <- intersect(gene_list, rownames(cluster_expr))
      # If at least one gene maps, calculate mean per cluster, else NA
      if (length(genes_in_expr) > 0) {
        colSums(cluster_expr[genes_in_expr, , drop = FALSE], na.rm = TRUE)
      } else {
        rep(NA, ncol(cluster_expr))
      }
    })
  ) %>%
  ungroup()

# Expand avg_expr into columns (one per cluster)
avg_matrix <- segment_cluster_averages %>%
  unnest_wider(avg_expr, names_sep = "_") # This will name columns avg_expr_ClusterName

# Optional: Rename columns to just cluster names
colnames(avg_matrix)[-(1:2)] <- colnames(cluster_expr)

# Result: avg_matrix contains segment, genes, and one column per cluster with averages
```

```r
library(dplyr)
library(tidyr)
cluster_names <- colnames(cluster_expr)
```

```r
segment_cluster_averages <- segment_genes %>%
  rowwise() %>%
  mutate(
    avg_expr = list({
      gene_list <- unlist(strsplit(genes, ","))
      genes_in_expr <- intersect(gene_list, rownames(cluster_expr))
      # Always return a named vector of length = ncol(cluster_expr)
      if (length(genes_in_expr) > 0) {
        avg <- colSums(cluster_expr[genes_in_expr, , drop = FALSE], na.rm = TRUE)
        # Ensure all clusters present, fill missing with 0
        avg[setdiff(cluster_names, names(avg))] <- 0
        avg <- avg[cluster_names] # correct order
      } else {
        setNames(rep(0, length(cluster_names)), cluster_names)
      }
    })
  ) %>%
  ungroup()

avg_matrix <- segment_cluster_averages %>%
  unnest_wider(avg_expr)
```
